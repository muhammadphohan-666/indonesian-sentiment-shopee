```python
from google_play_scraper import app, reviews, Sort
from google.colab import files
import pandas as pd
```

```python
APP_ID = "com.shopee.id"

try:
    info = app(APP_ID, lang="id", country="id")
    print("App ditemukan!")
    print("Title:", info.get("title"))
    print("AppId:", info.get("appId"))
except Exception as e:
    print("AppId tidak valid / tidak ditemukan:", e)
```

```python
APP_ID = "com.shopee.id"
COUNT = 40000
LANG = "id"
COUNTRY = "id"

result, _ = reviews(
    APP_ID,
    lang=LANG,
    country=COUNTRY,
    sort=Sort.NEWEST,
    count=COUNT
)

df = pd.DataFrame(result)
print("Jumlah review keambil:", len(df))

df.to_csv("shopee_mobile_reviews.csv", index=False, encoding="utf-8-sig")
print("Selesai")
```
